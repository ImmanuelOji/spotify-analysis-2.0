```python
imm_stream.head()
```

```python
## to get the time frame of the song played in miliseconds
vic_stream['Time-Played (hh-mm-ss)'] = pd.to_timedelta(vic_stream['msPlayed'], unit = 'ms')
pet_stream['Time-Played (hh-mm-ss)'] = pd.to_timedelta(pet_stream['msPlayed'], unit = 'ms')
ned_stream['Time-Played (hh-mm-ss)'] = pd.to_timedelta(ned_stream['msPlayed'], unit = 'ms')
imm_stream['Time-Played (hh-mm-ss)'] = pd.to_timedelta(imm_stream['msPlayed'], unit = 'ms')
```

```python
vic_stream.head()
```

```python
def hours(td):
    '''To get the hour infomation'''
    return td.seconds/3600

def minutes(td):
    '''To get the minutes information'''
    return (td.seconds/60)%60

vic_stream['Listening Time(Hours)'] = vic_stream['Time-Played (hh-mm-ss)'].apply (hours).round(3)
vic_stream['Listening Time(Minutes)'] = vic_stream['Time-Played (hh-mm-ss)'].apply (minutes). round(3)
pet_stream['Listening Time(Hours)'] = pet_stream['Time-Played (hh-mm-ss)'].apply (hours).round(3)
pet_stream['Listening Time(Minutes)'] = pet_stream['Time-Played (hh-mm-ss)'].apply (minutes). round(3)
ned_stream['Listening Time(Hours)'] = ned_stream['Time-Played (hh-mm-ss)'].apply (hours).round(3)
ned_stream['Listening Time(Minutes)'] = ned_stream['Time-Played (hh-mm-ss)'].apply (minutes). round(3)
imm_stream['Listening Time(Hours)'] = imm_stream['Time-Played (hh-mm-ss)'].apply (hours).round(3)
imm_stream['Listening Time(Minutes)'] = imm_stream['Time-Played (hh-mm-ss)'].apply (minutes). round(3)
```

```python
pet_stream.head()
```

```python
## dropping colums we don't need
vic_stream.drop(columns=['endTime', 'msPlayed', 'Time-Played (hh-mm-ss)'], inplace = True)
pet_stream.drop(columns=['endTime', 'msPlayed', 'Time-Played (hh-mm-ss)'], inplace = True)
ned_stream.drop(columns=['endTime', 'msPlayed', 'Time-Played (hh-mm-ss)'], inplace = True)
imm_stream.drop(columns=['endTime', 'msPlayed', 'Time-Played (hh-mm-ss)'], inplace = True)
```
